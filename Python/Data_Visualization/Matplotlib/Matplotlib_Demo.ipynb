```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r.')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Plot_1')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Plot_2')
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1.0, 1.0])
axes.plot(x, y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('OO Plot')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x, y, 'r')
ax2.plot(x, y, 'g')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3)
#plt.tight_layout()
for current_ax in ax:
    current_ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize=(8, 4), nrows=2, ncols=1)

ax[0].plot(x, y)
ax[1].plot(y, x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(y, x, label="x y", linewidth=4, alpha=0.5, marker='*', markersize=14)
ax.plot(x, y, label="y x", lw=4, linestyle='--', marker='o', markersize=12, 
        markerfacecolor='purple', markeredgewidth='3', markeredgecolor='g')
ax.legend()
```

```python
plt.scatter(x, y)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```
