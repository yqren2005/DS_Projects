```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.distplot(tips.total_bill, bins=40)
```

```python
sns.jointplot('total_bill', 'tip', tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips.total_bill)
```

```python
sns.kdeplot(tips.total_bill)
```

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
import numpy as np
sns.barplot(x='sex', y='tip', data=tips, estimator=np.std)
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
sns.heatmap(flights.corr(), annot=True, cmap='coolwarm')
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights.pivot_table(index='month', columns='year', values='passengers'), 
            linecolor='w', linewidths=0.25, cmap='coolwarm')
```

```python
flights.shape
```

```python
sns.clustermap(flights.pivot_table(index='month', columns='year', values='passengers'), cmap='coolwarm', standard_scale=1)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['o', 'v'], scatter_kws={'s':60})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['o', 'v'], scatter_kws={'s':80}, col='sex', row='time')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
sns.pairplot(iris, hue='species')
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker', hue='sex')
g.map(plt.scatter, 'total_bill', 'tip')
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='seismic', hue='smoker')
sns.despine()
```
