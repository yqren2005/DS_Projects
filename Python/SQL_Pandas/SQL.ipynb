```python
import sqlite3
```

```python
import pandas as pd
```

```python
con = sqlite3.connect('sakila.db')
```

```python
query = 'SELECT * from customer'

df = pd.read_sql(query, con)

df
```

```python
def sql_to_df(q):
    df = pd.read_sql(q, con)
    return df
```

```python
query = 'select first_name, last_name from customer'
df = sql_to_df(query)
df.head()
```

```python
query = 'select distinct country_id from city;'
sql_to_df(query).shape
```

```python
sql_to_df(query).tail()
```

```python
query = 'select * from customer where store_id = 1'
sql_to_df(query).head()
```

```python
query = "select * from customer where first_name = 'MARY'"
sql_to_df(query)
```

```python
query = "select * from film where release_year = 2006 and rating = 'R'"
sql_to_df(query).head()
```

```python
query = "select * from film where rating = 'PG' or rating = 'R'"
sql_to_df(query).head()
```

```python
query = "select count(customer_id) from customer"
sql_to_df(query).head()
```

```python
query = "select * from customer where last_name like 'M%'"
sql_to_df(query).head()
```

```python
query = "select * from customer where last_name like '____ing'"
sql_to_df(query).head()
```

```python
# in mysql use: select * from customer where last_name like '[AB]%'

query = "select * from customer where last_name glob '[AB]*' and first_name like '%A'"
sql_to_df(query).head()
```

```python
query = "select * from customer where last_name glob '[AB]*' and first_name like '%A' order by last_name DESC"
sql_to_df(query)
```

```python
query = 'select store_id, count(customer_id) from customer group by store_id'
sql_to_df(query).head()
```

```python
query = "select first_name, last_name, last_update, count(customer_id) from customer where last_name glob '[AB]*' and first_name like '%A' group by last_update order by first_name DESC"
sql_to_df(query)
```
